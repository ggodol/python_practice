##다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

- 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

    >>> addall([1])
    1
    >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
    45

```python
def addall_for(list): #for ~ in 리터널방식
    result = 0
    for x in list:
        result += x
    return result

def addall_import(list): #리스트 내포방식
    return sum([x for x in list])
    
#for ~ in 리터널방식
print addall_for([1])
print addall_for([1,2,3,4,5,6,7,8,9])

#리스트 내포방식
print addall_import([1])
print addall_import([1,2,3,4,5,6,7,8,9])
```

- 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

    >>> addallodd([1])
    1
    >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
    25

[참고]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)

```python
def addallodd_for(list): #for ~ in 리터널방식
    result = 0
    for x in list:
        if(x % 2== 1):
            result += x
    return result

def addallodd_import(list): #리스트 내포방식
    return sum([x for x in list if x % 2 == 1])
    
#for ~ in 리터널방식
print addallodd_for([1])
print addallodd_for([1,2,3,4,5,6,7,8,9])

#리스트 내포방식
print addallodd_import([1])
print addallodd_import([1,2,3,4,5,6,7,8,9])
```

##[프로젝트 오일러 문제 7]

소수를 크기 순으로 나열하면 2, 3, 5, 7, 11, 13, ... 과 같이 됩니다.

이 때 10,001번째의 소수를 구하세요.

```python
def isPrime(number): #number가 소수이면 True를 리턴하는 함수
    if number == 2:
        return True
    for x in range(2,number):
        if number % x == 0:
            break
        elif x == number - 1:
            return True
          
num = 2
order = 1

while (order <= 10001):
    if(isPrime(num)==True):
        order+=1
    num+=1
    
print num-1 #while문 마지막에 num+1을 하기때문에 1을 뺀 수가 정답
```

isPrime은 들어오는 인자가 소수이면 True를 반환하는 함수이다. while문을 돌면서 숫자를 하나씩 소수인지 검사하고, order에 몇번째 소수인지 기록한다. 10001번째 소수가 나올 때까지 while문이 돌면, 나오는 num-1이 정답이다. 1을 빼주는 이유는 while문 안에서 마지막 num을 1 더해주기 때문이다.

##[프로젝트 오일러 문제 9]

세 자연수 a, b, c 가 피타고라스 정리 a2 + b2 = c2 를 만족하면 피타고라스 수라고 부릅니다 (여기서 a < b < c ).
예를 들면 32 + 42 = 9 + 16 = 25 = 52이므로 3, 4, 5는 피타고라스 수입니다.

a + b + c = 1000 인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
a = 1
b = 2
sum = 1000
flag = 0 #break를 하기위한 flag

while (a < b):
    b = a+1
    for x in range(a,(sum-a-1)/2): #b는 a보다 크고, c보다 작은 범위이다
        c = sum-(a+b)
        if(a*a + b*b == c*c): #피타고라스 수를 만족시키면 flag를 1로 바꾸고 break
            flag = 1
            break
        b+=1
    
    if(flag == 1): #조건을 만족하는 피타고라스 수를 만났으니 while문을 빠져나온다
        break
    a += 1

print a*b*c
```

a가 1부터, b가 2부터 하나씩 증가하다가 피타고라스 수를 만족하면 break하는 방법을 사용했다. while은 b가 a보다 클때만 수행되며, a가 1씩 더해진다. for문은 b가 a보다 크고, c보다 작은 범위에서 수행되며, b가 1씩 더해진다. c는 a와 b를 알면 a+b+c=1000에 의해서 자동으로 구해진다. 지속적으로 while문과 for문이 수행되다보면 피타고라스 수를 만족하는 a,b,c 값이 나오게된다. 만족하는 값이 나오게되면 for문에서 break로 나오게 되며, flag를 0에서 1로 바꾸어 while문도 나오게 된다. 결과적으로 모든 조건을 만족시키는 a,b,c를 구할 수 있다.

###[Incremental Project] 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.

- 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
    - 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
    - 모든 punctuation을 확인하는 방법은 아래와 같다.

    import string
    print string.punctuation

- 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
    - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']

- 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
    - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}

- 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
    - http://www.naver.com
    - http://www.daum.net
    - http://www.nytimes.com

```python
import urllib2
import string
import pprint

class MyPrettyPrinter(pprint.PrettyPrinter): #프린트한 결과가 한글로 나오게하기 위한 class(참고: http://egloos.zum.com/mcchae/v/11076302)
    def format(self, _object, context, maxlevels, level):
        if isinstance(_object, unicode):
            return "'%s'" % _object.encode('utf8'), True, False
        elif isinstance(_object, str):
            _object = unicode(_object,'utf8')
            return "'%s'" % _object.encode('utf8'), True, False
        return pprint.PrettyPrinter.format(self, _object, context, maxlevels, level)

def extraction_dictionary(source):#주소를 적으면 해당 사이트 단어들의 출현 빈도를 사전형태로 리턴
    source = urllib2.urlopen(source).read() #source로 소스코드를 모두 문자열로 가져옴

    for i in range(0, source.count('<!--')): #주석태그부분 처리
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    for i in range(0,source.count("<")): #태그부분 처리
        if source.find("<") != -1:
            source = source.replace(source[source.find("<"):source.find(">")+1:],"")

    words = source.split() #words 리스트에 단어추출

    pure_words = []

    for s in words: #words 원소의 문자열을 하나하나씩 꺼내어 구두문자 제거
        for x in s: #문자열의 문자 하나하나씩 꺼내서 구두문자인지 비교
            if(x in string.punctuation): #문자가 구두문자이면 그 문자 제거
                s = s[:s.index(x)]+s[s.index(x)+1:]
        if(s != ''): #구두문자를 제거했는데 빈문자열이면 리스트에 추가하지 않는다
            pure_words.append(s)

    D = {} #출현빈도를 저장할 사전

    for x in pure_words:
        if(x in D): #단어가 key값에 이미 존재하면 value값에 1을 더한다.
            D[x] += 1
        else: #단어가 key값에 존재하지 않으면 value값을 1로하여 key를 추가한다.
            D[x] = 1
    return D

print "http://www.naver.com 결과"
MyPrettyPrinter().pprint(extraction_dictionary("http://www.naver.com")) #프린트한 결과의 문자가 한글로 잘 출력하기 위한 메소드 수행
print MyPrettyPrinter().pformat(extraction_dictionary("http://www.naver.com"))

print "http://www.daum.net 결과"
MyPrettyPrinter().pprint(extraction_dictionary("http://www.daum.net")) #프린트한 결과의 문자가 한글로 잘 출력하기 위한 메소드 수행
print MyPrettyPrinter().pformat(extraction_dictionary("http://www.daum.net"))

print "http://www.nytimes.com 결과"
MyPrettyPrinter().pprint(extraction_dictionary("http://www.nytimes.com")) #프린트한 결과의 문자가 한글로 잘 출력하기 위한 메소드 수행
print MyPrettyPrinter().pformat(extraction_dictionary("http://www.nytimes.com"))
```
