다형성을 간단히 4단어로 말하면 이렇게 표현할 수 있다. same interface, different behavior. 풀어서 말하면, 같은 상속관계 내의 클래스들의 인스턴스들의 메소드가 같은 이름이지만, 다른 기능을 한다는 것이다. 이 것으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있다.

```python
class nation:
    def hello(self):
        print 'hello'
    
class korea(nation):
    def hello(self):
        print '안녕하세요.'

class japan(nation):
    def hello(self):
        print 'konizziwa'
        
class china(nation):
    def hello(self):
        print 'nihao'
        
class vietnam(nation):
    def hello(self):
        print 'ssinzao'
        
for nation in (korea(), japan(), china(), vietnam()):
    nation.hello()
```

***[3번문제]***  
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
    
> **\>>> c = Counter(10)**  
> **\>>> d = Counter(10, 2)**
        
- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \__str__\() 메소드를 Counter 클래스 내에 구현하시오.

> **\>>> print c**  
> **[Count (step: 1)] 10**  
> **\>>> print d**  
> **[Count (step: 2)] 10**
     
- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

> **\>>> c.incr()**  
> **\>>> d.incr()**  
> **\>>> print c**  
> **\[Count (step: 1)] 11**   
> **\>>> print d**  
> **\[Count (step: 2)] 12**  

- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

> **\>>> c()**  
> **\>>> d()**  
> **\>>> print c**   
> **\[Count (step: 1)] 12**   
> **\>>> print d**   
> **\[Count (step: 2)] 14** 

- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> **\>>> c = c + 5**     
> **\>>> d = d - 5**   
> **\>>> print c**   
> **[Count (step: 1)] 17**   
> **\>>> print d**  
> **[Count (step: 2)] 9** 

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> **\>>> print c > 10**  
> **True**  
> **\>>> print d > 10**  
> **False**  
> **\>>> print c < 10**  
> **False**  
> **\>>> print d < 10**  
> **True**  
> **\>>> print c == 17**   
> **True**  
> **\>>> print d != 9**  
> **False**

```python
class Counter:
    def __init__(self, count, step = 1): #요구사항1
        self.count = count
        self.step = step

    def __str__(self): #요구사항2
        return "[Count (step:%d)]%d" %(self.step,self.count)

    def incr(self): #요구사항3
        self.count += self.step 

    def __call__(self): #요구사항4
        self.incr()
    
    def __add__(self, other): #요구사항5
        self.count += other
        return self
    
    def __sub__(self, other): #요구사항5
        self.count -= other
        return self
    
    def __gt__(self, other): #요구사항6
        return self.count > other
    
    def __lt__(self, other): #요구사항6
        return self.count < other
    
    def __eq__(self, other): #요구사항6
        return self.count == other
    
    def __ne__(self, other): #요구사항6
        return self.count != other

c = Counter(10)
d = Counter(10,2)

print c
print d

c.incr()
d.incr()

print c
print d

c()
d()

print c
print d

c=c+5
d=d-5

print c
print d

print c>10
print d>10
print c<10
print d<10
print c==17
print d!=9
```

***[4번문제]***  
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

MySet의 역할부터 설명하면, MySet은 집합(Set)자료형을 만들기위한 클래스이다. 여기서 집합(Set)자료형은, 집합과 같이 같은 원소가 두 번 존재하지 않는 자료형이다. 

\_\_init\_\_()메소드를 살펴보자. 역할은 list를 받아서, 중복된 원소는 하나만 남게 제거하는 생성자 메소드이다. MySet 클래스는 내장 자료형 list를 상속받았다. 따라서, 인자로 list를 받는다. for문 안에서, 인자로 받은 list에서 원소를 하나씩 빼서, 생성된 인스턴스 객체의 리스트에 하나씩 추가한다. 후에, MySet클래스의 정적메소드 elimicate_duplicate()를 생성하여 원소가 하나씩만 존재하도록, 중복된 원소를 제거한다. elimicate_dupicate()메소드는 아래에서 자세히 설명하겠다.

\_\_str\_\_()메소드의 역할은 MySet 인스턴스로 생성된 리스트를 집합형태로 출력하는 것이다. result는 결과적으로 출력할 문자열을 저장한다. "MySet: {1 ,2 ,3 ,4 ,5}" 형태로 출력한다.

eliminate_duplicate()메소드는 리스트 내에서 원소들이 하나만 남도록 중복된 원소를 제거하는 역할을 한다. @staticmethod가 위에 적혀있는데, 이것은 이 메소드는 정적메소드라는 것이다. 따라서 인스턴스 객체와 무관하게 클래스 이름을 이용하여 직접 호출할 수 있다. s=[]를 통해, s라는 임시리스트를 생성한다. 그리고, 인자로 받은 리스트 l에서 하나씩 꺼내서 s로 집어넣는다. 이때, if e not in s 조건문을 통해서 이미 s리스트에 있는 원소는 더이상 추가하지 않는다. 따라서 결과적으로는 s리스트에는 l리스트의 원소를 종류별로 하나씩만 가져오게된다. l[:]=[]을 통해서 l리스트를 깨끗하게 비운다. 그리고 다시 l리스트에 s리스트의 존재하는 원소를 모두 옮긴다. 최종결과로는 l리스트에는 중복되는 원소가 없게 된다.

***[5번문제]***  
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> **\>>> u = s | t**  
> **\>>> print u**  
> **MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}**

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> **\>>> u = s & t**   
> **\>>> print u**  
> **MySet: {2, 3}**

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> **\>>> s = MySet([1, 2, 3])**  
> **\>>> t = MySet([3, 4, 5])**  
> **\>>> u = s - t**  
> **\>>> print u**  
> **MySet: {1}**

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, other):
        result = []
        result.extend(other)
        result.extend(self)
        MySet.eliminate_duplicate(result)
        result.sort()
        return MySet(result)

    def __and__(self, other):
        temp = []
        result = []
        temp = self | other
        for e in temp:
            if(e in self and e in other):
                result.append(e)
        return MySet(result)


    def __sub__(self, other):
        result = [] 
        result.extend(self)
        for e in other:
            if(e in result):
                result.remove(e)
        return MySet(result)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t

    u=s|t
    print u

    u = s&t
    print u

    s = MySet([1,2,3])
    t = MySet([2,3,4])

    u = s-t
    print u
```

####[프로젝트 오일러 문제 17] 

1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고,
각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.
  예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,
  115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.

```python
# -*- coding: utf-8 -*-
numdict={1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight'
, 9:'nine', 10:'ten', 11:'eveven', 12:'twelve', 13:'thirteen', 14:'fourteen', 15:'fifteen'
, 16:'sixteen', 17:'seventeen', 18:'eighteen', 19:'nineteen', 20:'twenty', 30:'thirty', 40:'forty'
, 50:'fifty', 60:'sixty', 70:'seventy', 80:'eighty', 90:'ninety', 100:'hundred', 1000:'thousand'}

#숫자를 넣으면 문장으로 바꾸는 함수
def inverter(num):
    string='' #반환할 문자열
    
    if(num < 21): #21보다 작을 때는 사전에 있는 내용 그대로 반환
        return numdict[num]

    elif(num < 100): #100보다 작을 때
        string += numdict[(num/10)*10]
        
        if((num%10) != 0): #일의 자리가 0이 아니면, 일의 자리도 문자열에 추가
            string+=numdict[num%10]
        
        return string
        
    elif(num < 1000): #백의 자리 숫자 일 때
        string += numdict[(num/100)]+'hundred' #백의 자릿 수를 나타내는 문자열과 hundred를 추가
        num = num-((num/100)*100) #num에서 100의 자릿 수 제거
        
        if(num != 0): #num이 0이아닐 경우, 뒤에 숫자가 존재함으로 string에 and를 추가한다.
            string +='and'
            
            if(num < 21): #21보다 작을 경우, 해당 숫자 사전에 꺼내서 바로 추가
                string+=numdict[num]
            else:
                string += numdict[(num/10)*10] #십의 자리 수를 나타내는 문자열 string에 추가
                if((num%10) != 0): #일의 자리가 0이 아니면, 일의 자리도 문자열에 추가
                    string+=numdict[num%10]
            
        return string
    
    elif(num==1000): #1000일 때
        return 'one'+numdict[num]
       
#총 문자 갯수를 저장할 변수
result = 0
 
for x in range(1,1001): #1부터 1000까지 문자열을 세서 더함
    result += len(inverter(x))

print result
```

numdict에 필요한 최소한의 숫자에 대한 문자를 추가한다. inverter 함수는 숫자를 인자로 받아서, 인자로 받은 숫자를 문자로 반환하는 함수이다. 최종적으로 for문으로 1부터 1000까지 숫자를 문자열로 바꿔서 문자열의 갯수를 result에 더한다. 순수 문자열을 계산 하기위하여, 띄어쓰기나 '-'은 생략했다.

다음은 달력에 관한 몇 가지 일반적인 정보입니다 (필요한 경우 좀 더 연구를 해 보셔도 좋습니다).

- 1900년 1월 1일은 월요일이다.
- 4월, 6월, 9월, 11월은 30일까지 있고, 1월, 3월, 5월, 7월, 8월, 10월, 12월은 31일까지 있다.
- 2월은 28일이지만, 윤년에는 29일까지 있다.
- 윤년은 연도를 4로 나누어 떨어지는 해를 말한다. 하지만 400으로 나누어 떨어지지 않는 매 100년째는 윤년이 아니며, 400으로 나누어 떨어지면 윤년이다

20세기 (1901년 1월 1일 ~ 2000년 12월 31일) 에서, 매월 1일이 일요일인 경우는 총 몇 번입니까?

```python
#해당 월의 날 수를 사전으로 저장
monthdays = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

#윤년일 경우 해당 월의 날 수를 사전으로 저장
leapdays = {1:31, 2:29, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

#해당연도(year)가 윤년인지 아닌지 판별하는 함수
def isLeapYear(year):
    if(year%100 == 0 and year%400 != 0): #400으로 나누어지지 않는 100년째는 윤년아님
        return False
    elif(year%4 != 0): #4로 나누어지지 않으면 윤년아님
        return False
    
    return True #그외의 경우는 모두 윤년

def findDay(year,month,day): #년, 월, 일을 입력 시 해당 날짜의 요일을 출력
    daysum = 0 #날짜의 총합을 저장할 변수
    
    
    #윤년이면 366일, 평년이면 365일을 daysum에 더한다.
    for y in range(1900,year):
        if(isLeapYear(y)==True):
            daysum+=366
        else:
            daysum+=365
          
    #윤년이면 2월이 29일, 평년이면 2월이 28일로 계산하여 daysum에 더한다.
    for d in range(1,month):
        if(isLeapYear(year)==True):
            daysum+=leapdays[d]
        else:
            daysum+=monthdays[d]
    
    daysum+=day #최종적으로 날짜를 더한다.
    return daysum%7 #7로 나눈다. 0이면 일요일, 1이면 월요일... 6이면 토요일이다.

result = 0 #결과값을 저장할 변수

#1901년 1월 1일부터, 2000년 12월 31일까지 중, 1일이 일요일(findDay=0)인 갯수를 찾는다
for y in range(1901,2001):
    for m in range(1,13):
        if(findDay(y,m,1)==0):
            result+=1

print result
```

findDay에 년, 월, 일을 넣으면 입력 받은 날의 요일을 구하는 함수를 구현했다. 반환값이 0이면 일요일, 1이면 월요일, 2이면 화요일... 6이면 토요일이다. 이 함수에서 윤년여부를 따지기 위하여 isLeapYear 함수를 만들었다. 윤년이면 True를 반환한다. 

for문을 통해서 1901년 1월 1일부터, 2000년 12월 31일까지 각 년도, 달의 1일을 findDay에 넣었다. 만약에 일요일이라면 1을 반환한다. 이 원리를 이용하여 결과를 얻을 수 있었다.

n! 이라는 표기법은 n × (n − 1) × ... × 3 × 2 × 1을 뜻합니다.

예를 들자면 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800 이 되는데,
여기서 10!의 각 자리수를 더해 보면 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27 입니다.

100! 의 자리수를 모두 더하면 얼마입니까?

```python
#factorial을 재귀적으로 구하는 함수
def factorial(num):
    if (num == 1):
        return 1
    else:
        return num*factorial(num-1)

num = 100 #fatorial에 넣을 자연수
result = 0 #각 자리수를 모두 더한 수

#factorial 연산 후, 나온 숫자를 문자열로 바꾼다. 
#그리고 문자 하나씩 꺼내여 int형으로 바꾸어 더한다.
for x in str(factorial(num)):
    result+=int(x)

print result
```

factorial을 재귀적으로 계산하는 함수 만들었다. factorial 계산으로 나온 결과 값을 문자열로 바꿔준다. 문자열로 바꾸는 이유는 문자열 하나하나가 자릿 수이기 때문에 연산하기 편해지기 때문이다. 따라서 for문을 통해서 하나하나의 자릿 수를 int형으로 변환하여 결과값에 더해준다.

####[프로젝트 오일러 문제 21]

n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,
서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 
a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
따라서 220과 284는 친화쌍이 됩니다.

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
import math

#약수를 리스트로 반환하는 함수
def getPrime(num):
    prime_list=[]
    
    #루트 num까지만 반복한다. 나눠지는 수를 원래 수로 나눈 수도 약수이다.
    #따라서 이 방법으로 하면, 좀 더 수행을 간단하게 할 수 있다. 
    for x in range(1,int(math.sqrt(num))+1):
        if(num%x == 0):
            prime_list.append(x)
            if(x*x != num):
                prime_list.append(num/x)
    
    return prime_list
    
#자신을 제외한 약수의 합을 구하는 함수
def d(num):
    result = 0
    prime_list = getPrime(num)
    
    for x in prime_list:
        result += x
        
    result -= num #자기자신을 뺀다.
    
    return result
    
#친화수를 저장하는 리스트
amicable_num_list = []

#10000이하의 수중 친화수를 찾아서 amicable_num_list에 저장
for x in range(1,10001):
    y = d(x)
    
    #d(x)=y이고 d(y)=x를 만족하고, x와 y가 같지않으면 친화수이다.
    if (d(x)==y and d(y)==x and x != y):
        if(x not in amicable_num_list):
            amicable_num_list.append(x) 
        
        if(y not in amicable_num_list):
            amicable_num_list.append(y)

print sum(amicable_num_list) #친화수 리스트에 있는 모든 친화수를 더한다
```

getPrime(num)은 num의 약수를 리스트로 반환하는 함수이다. d(num)은 num의 약수 중 자신을 제외한 합을 구하는 함수이다. d(num)함수를 이용하여, 친화쌍을 구한다. 구해진 친화쌍을 친화쌍을 저장하는 리스트인 amicable_num_list에 저장한다. 최종적으로 친화수리스트에 있는 모든 친화수를 더하여 결과를 출력한다.

```python
# -*- coding: utf-8 -*-
import urllib2
import string

class WebWordsFrequency():
    def __init__(self, *args): #가변인수로 URL을 가져와 리스트에 저장한다.
        self.sources = []
        for source in args:
            self.sources.append(source)

    def addUrl(self, source): #URL을 하나 추가
        self.sources.append(source)

    def removeUrl(self, source): #URL을 삭제
        if source in self.sources:
            self.sources.remove(source)

    def listUrls(self): #현재 등록된 모든 URL을 출력
        for source in self.sources:
            print source

    def getWordsFrequency(self): #각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환
        D = {} #사전에 있는 단어들을 하나로 통합할 사전
        item_list = []

        #하나로 통합하기 위해서 사전을 리스트로 반환
        for source in self.sources:
            item_list.append(self.extraction_dictionary(source).items())

        #리스트에서 (words,values)로 되어있는 튜플을 하나씩 확인하여 단어 빈도를 D 사전에 저장
        for x in item_list: #리스트에서 한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트 하나씩 꺼냄 
            for words in x: #한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트에서 하나의 튜플을 꺼냄
                if(words[0] in D): #(단어, 빈도)로 저장되어있는 튜플 words에서 단어가 D 사전에 이미 존재하면 빈도를 +1
                    D[words[0]] += 1
                else: #단어가 존재하지 않으면 빈도를 1로하여 추가한다.
                    D[words[0]] = 1
        return D

    def  getMaxFreqencyWords(self): #각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전에서 가장 많이 출현한 단어 리스트를 반환
        D = self.getWordsFrequency()

        #sort하기 위하여 D 사전을 키와 값의 튜플 리스트로 변환
        item_list = D.items()
        
        if(len(item_list)!=0): #등록된 URL이 하나라도 있으면
            #단어의 빈도(사전의 value)값으로 정렬
            item_list.sort(key=lambda t:t[1], reverse=True)
    
            x = 0 
            while(item_list[x][1]==item_list[0][1]): #정렬된 튜플의 처음 key값과 같은 key 값을 가진 단어 모두 출력
                print item_list[x][0]
                x+=1
        else: #등록된 URL이 하나도 없으면, None 반환
            return None
        

    @staticmethod
    def extraction_dictionary(source):#주소를 적으면 해당 사이트 단어들의 출현 빈도를 사전형태로 리턴
        source = urllib2.urlopen(source).read() #source로 소스코드를 모두 문자열로 가져옴
        #영어 불용어 리스트
        stopword = [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']

        for i in range(0, source.count('<!--')): #주석태그부분 처리
            if source.find("<!--") != -1:
                source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

        for i in range(0,source.count("<")): #태그부분 처리
            if source.find("<") != -1:
                source = source.replace(source[source.find("<"):source.find(">")+1:],"")

        words = source.split() #words 리스트에 단어추출

        pure_words = []

        for s in words: #words 원소의 문자열을 하나하나씩 꺼내어 구두문자 제거
            for x in s: #문자열의 문자 하나하나씩 꺼내서 구두문자인지 비교
                if(x in string.punctuation): #문자가 구두문자이면 그 문자 제거
                    s = s[:s.index(x)]+s[s.index(x)+1:]
            
            #구두문자를 제거했는데 빈문자열이면 리스트에 추가하지 않는다
            #영어 불용어면 추가하지 않는다
            if(s != '' and s not in stopword): 
                pure_words.append(s)


        D = {} #출현빈도를 저장할 사전

        for x in pure_words: #구두문자를 제거한 단어 하나하나씩 꺼내여 사전에 저장
            if(x in D): #단어가 key값에 이미 존재하면 value값에 1을 더한다.
                D[x] += 1
            else: #단어가 key값에 존재하지 않으면 value값을 1로하여 key를 추가한다.
                D[x] = 1
        return D



w1 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com') 
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com') 
w3 = WebWordsFrequency()

w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
w1.removeUrl('http://www.amazon.com')
w2.removeUrl('http://stackoverflow.com')
w1.listUrls()

print w2.getWordsFrequency()
w1.getMaxFreqencyWords()
```

13번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
- getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

    - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
    
> **\>>> w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')**  
>
> **\>>> w4.getWordsFrequency()  
[('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]**
>
> **\>>> w4.getWordsFrequency(reverse=true)  
[('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]**

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse = False): #각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환
        D = {} #사전에 있는 단어들을 하나로 통합할 사전
        item_list = []

        #하나로 통합하기 위해서 사전을 리스트로 반환
        for source in self.sources:
            item_list.append(self.extraction_dictionary(source).items())

        #리스트에서 (words,values)로 되어있는 튜플을 하나씩 확인하여 단어 빈도를 D 사전에 저장
        for x in item_list: #리스트에서 한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트 하나씩 꺼냄 
            for words in x: #한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트에서 하나의 튜플을 꺼냄
                if(words[0] in D): #(단어, 빈도)로 저장되어있는 튜플 words에서 단어가 D 사전에 이미 존재하면 빈도를 +1
                    D[words[0]] += 1
                else: #단어가 존재하지 않으면 빈도를 1로하여 추가한다.
                    D[words[0]] = 1

        #sort하기 위하여 D 사전을 키와 값의 튜플 리스트로 변환
        item_list = D.items()
        
        #리스트 sort메소드를 이용하기위해서, reverse값에 따라 tf를 조정한다.
        if(reverse == False):
            tf = True
        else:
            tf = False
        
        #단어의 빈도(사전의 value)값으로 정렬
        item_list.sort(key=lambda t:t[1], reverse=tf)
        
        print item_list
                    
        
w4 = OrderedWebWordsFrequency('http://www.times.com', 'http://www.cnn.com', 'https://github.com')
w4.getWordsFrequency()
w4.getWordsFrequency(reverse=True)
```
