과제 수행을 위한 파일을 생성한다.

- 1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.
>cat dog  
>dog pig  
>ham bird  
>pig ham

```python
import os

f1 = open('s.txt','r')
string_list = f1.read().splitlines() #s.txt를 라인마다 \n을 제거하여 리스트에 저장
f1.close()

temp_list = [] #하나의 문장을 단어별로 나누어 저장할 리스트

for s in string_list:
    temp_list.append(s.split()) #하나의 문장을 단어별로 만들어 temp_list 리스트에 추가

def my_key(a): #sort할 때, 몇 번째 단어를 기준으로 할 지 key값 결정 
    return a[0]

temp_list.sort(key = my_key) #단어 정렬

string = '' #s1.txt에 저장할 문자열 생성

for x in temp_list: #리스트에 저장된 단어하나하나 꺼내어 텍스트파일에 저장할 문자열 생성
    for y in x:
        string += y+' '
    string += '\n'

f2 = open('s1.txt','w')
f2.write(string) #s1.txt에 알파벳 순으로 정렬된 결과 값을 쓴다
f2.close()

print os.getcwd() #파일경로 출력
print string #s1.txt에 저장한 문자열 출력
```

>pig ham cat  
>dog ham bird

```python
import os

f1 = open('s.txt','r')
string = f1.read()
f1.close()

temp_list = string.split() #리스트에 띄어쓰기를 기준으로 낱말 저장

count = 0 #한 라인에 3단어만 출력하도록 카운트하는 변수
new_string = '' #s3.txt에 저장할 문자열

for s in temp_list: #리스트에서 단어 하나씩 꺼내서 문자열에 더한다.
    if(count == 3): #3개의 단어가 한 줄에 나열되면 개행한다.
        new_string += '\n'+s+' '
        count = 0
    else: #한 줄에 3개의 단어가 아직 안나왔으므로, 뒤에 단어를 추가한다.
        new_string += s+' '
    
    count += 1
        
print new_string #저장된 문자열 출력

f2 = open('s3.txt','w')
f2.write(new_string)
f2.close()
```

***[3번 문제]***  

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop)  
> range(start, stop[, step])
* 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
* 3) frange 사용 예
    * 3-1) frange(0.5)
        * [0.0, 0.1, 0.2, 0.3, 0.4]
    * 3-2) frange(1.0, 2.0)
        * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    * 3-3) frange(2.2, 4.0, 0.5)
        * [2.2, 2.7, 3.2, 3.7]

```python
def frange(*args):
    start = 0.0
    step = 0.1
    result = [] #결과 리스트
    
    if(len(args)==1): #인자가 하나만 들어올 경우(stop)
        x = 0
        stop = args[0]
        
        while(x<stop):
            result.append(round(start+x,2)) #소수 두번째 자리에서 반올림
            x+=step
            
        return result 
    
    elif(len(args)==2): #인자가 두 개 들어올 경우(start,stop)
        x = 0
        start = args[0] #첫 번째 인자가 start
        stop = args[1] #두 번째 인자가 stop
        
        while(start+x<stop):
            result.append(round(start+x,2)) #소수 두번째 자리에서 반올림
            x+=step
            
        return result      

    elif(len(args)==3): #인자가 세 개 들어올 경우(start,stop, step)
        x = 0
        start = args[0] #첫 번째 인자가 start
        stop = args[1] #두 번째 인자가 stop
        step = args[2] #세 번째 인자가 step
        
        while(start+x<stop):
            result.append(round(start+x,2)) #소수 두번째 자리에서 반올림
            x+=step
            
        return result           
            
    else: #인수가 없거나 4개이상 실행 불가능
        print u"인수 갯수를 확인하세요."

print frange(0.5)
print frange(1.0,2.0)
print frange(2.2,4.0,0.5)
```

가변인수를 이용하여 문제를 풀었다. 기본 인수 값을 이용해서 풀이하려고 했지만, 인수가 하나 일때는 stop을 받고, 두 개 일 때는 start, stop으로 받아야되기 때문에, 기본 값 설정이 맨 뒤에 무조건 가야함으로 하지 못했다. 따라서 *args 가변인수를 이용하여 풀이했다.

***[4번 문제]***  

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

* sum() 사용 예
    * sum()
        * 0
* sum(1, 2)
    * 3
* sum(1, 2, 3, 4, 5)
    * 15
* sum(1, 5, 7, 2, -10)
    * 5

```python
def sum(*args):
    sum = 0
    
    for x in args: #가변인수를 하나씩 꺼내서 더함
        sum += x
    
    return sum

print sum()
print sum(1,2)
print sum(1,2,3,4,5)
print sum(1,5,7,2,-10)
```

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

* [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if(n==1):
        return 1
    else:  
        return n * myfact(n-1)
        
print myfact(950)
```

참을 수 있는 정도 만큼의 수행시간을 기다리려고 했다. 그러나 Jupyter와 Canopy에서 실행한 결과 950 가까이 되는 수들은 모두 바로 수행되었다. 하지만 1000 가까이되자 RuntimeError : maximum recursion depth exceeded 에러가 떴다. 아마 파이썬 내부적으로 재귀함수를 빠르게 처리하는 방법이 존재하는 것 같다. 그래서 정해진만큼 이상의 재귀함수로 들어가면 에러를 출력하는 것 같다.

####[프로젝트 오일러 문제 10]  
10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.

이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
import math

primeList = [2] #소수 리스트 생성
num = 3 #3부터 소수인지 검사한다
result = 0 #소수의 총합을 저장할 변수

while (num <= 2000000): 
    for x in primeList:
        if(num % x == 0): #소수로 나누었을 때 나누어지면 소수아님
            break
        elif(x > math.sqrt(num)): #num의 루트n까지도 나누어지지 않으면 소수
            primeList.append(num) #구한 소수를 소수리스트에 추가
            break
    num += 1

for x in primeList:
    result += x
    
print result
```

두 가지 방법을 사용하여 연산시간을 줄였다.
1. N보다 작은 소수로 나누었을 때, 나누어지지 않으면 소수이다.
2. N을 2부터 sqrt(N)까지 나눠보아 나누어지지 않으면 N은 소수이다.

primeList에는 소수를 저장하고 이 리스트에 있는 소수로만 나누었다. 그리고 나누는 수가 sqrt(N)보다 작은 수로만 나눠서 소수를 구하였다.

####[프로젝트 오일러 문제 12]
1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.
예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.
이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...  
이 삼각수들의 약수를 구해봅시다.

 **1**: 1  
 **3**: 1, 3  
 **6**: 1, 2, 3, 6  
**10**: 1, 2, 5, 10  
**15**: 1, 3, 5, 15  
**21**: 1, 3, 7, 21  
**28**: 1, 2, 4, 7, 14, 28  
위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
import math

divisor_list=[1] #약수를 저장하는 리스트
n = 1
k = 1

while(len(divisor_list) <= 500):
    divisor_list=[1] #약수리스트 초기화

    if(math.sqrt(n)==int(math.sqrt(n))): #sqrt(n)이 약수이면 저장
        divisor_list.append(int(math.sqrt(n))) 
            
    for x in range(2, int(math.sqrt(n))+1): #sqrt(n)보다 작은 n까지 나눠서 약수인지 확인
        if(n % x == 0): #약수이면
            divisor_list.append(x) #0으로 나누어 떨어지는 수와
            divisor_list.append(n/x) #n을 x로 나눈 수도 약수이기 때문에 약수리스트에 추가한다
            
    divisor_list.append(n) #자기자신을 리스트에 추가
    k += 1
    n = k*(k+1)/2 #다음 삼각수를 구한다

print divisor_list[-1]
```

무식하게 약수를 구하면 쉽게 구할 수 있지만, 삼각수를 구하고 약수를 구하는 알고리즘의 효율을 높이려고 노력했다. 삼각수를 구하는 방법은 등차수열의 합을 생각했다. n(n+1)/2 공식을 이용하면, n이 1씩 증가하면 다음 삼각수를 구할 수 있다. 

약수를 구하는 방식은 약수를 구하려는 수를 n이라 할 때, sqrt(n)까지만 나눠보는 방식을 이용했다. 왜냐하면 sqrt(n)보다 작은 약수는 n으로 나누면 sqrt(n)보다 큰 약수를 바로 구할 수 있기 때문이다. 여기에 sqrt(n)이 자연수면 약수이므로 약수리스트에 추가해주면 된다.

####[프로젝트 오일러 문제 14]
양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

n → n / 2 (n이 짝수일 때)  
n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

**참고:** 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
longpath = [0,0] #리스트 첫 번째는 숫자, 두 번째는 과정이 몇 번(count)인지 저장

for number in range(1,1000000): #1부터 백만까지 수행
    count = 0 #몇 번의 과정을 거쳐서 1이 되었는지 저장
    n = number #while문에서 규칙에 따라 1이되는 임의의 수
    
    while(n != 1): #주어진 규칙에 따라 1이되는 과정
        if(n%2==0): #n가 짝수 일 때
            n/=2
        else: #n가 홀수 일 때
            n = 3*n+1
        count += 1
    
    if(longpath[1] < count): #기존의 가장 긴 과정보다 더 길면, 값을 바꾼다
        longpath[0] = number
        longpath[1] = count

print longpath[0] 
```

규칙 그대로 풀었다. for문으로 1부터 백만까지 수행한다. 안에 while문은 주어진 규칙에 따라 1까지 도달한다. 그리고 아래에 if문으로 가장 긴 과정을 비교하여 저장하는 방법을 사용했다.

![그림](http://euler.synap.co.kr/images/p_015.gif)

그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?

```python
import math

width = 20 #가로
height = 20 #세로

#순열의 개념을 이용
number_of_route = math.factorial(width+height)/(math.factorial(width)*math.factorial(height))

print number_of_route
```

위에 예제를 보면 가로의 길이 2개, 세로의 길이 2개이다. 가로의 길이가 1인 경로를 a, 세로의 길이가 1인 경로를 b라 하자. 모든 경로는 가로 경로 a,a와 세로 경로 b,b를 순서대로 줄 세우는 것과 같은 개념이다. 위의 경우 (a,a,b,b), (a,b,a,b), (a,b,b,a), (b,b,a,a), (b,a,b,a), (b,a,a,b)로 6개의 방법으로 줄세울 수 있다. 따라서 순열의 개념으로 생각해서 풀었다.

* 5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들에서 가장 많이 출현한 단어 3개를 뽑아 제시하시오.
    * 반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함

```python
# -*- coding: utf-8 -*-
import urllib2
import string
import pprint
import pickle

class MyPrettyPrinter(pprint.PrettyPrinter): #프린트한 결과가 한글로 나오게하기 위한 class(참고: http://egloos.zum.com/mcchae/v/11076302)
    def format(self, _object, context, maxlevels, level):
        if isinstance(_object, unicode):
            return "'%s'" % _object.encode('utf8'), True, False
        elif isinstance(_object, str):
            _object = unicode(_object,'utf8')
            return "'%s'" % _object.encode('utf8'), True, False
        return pprint.PrettyPrinter.format(self, _object, context, maxlevels, level)

def extraction_dictionary(source):#주소를 적으면 해당 사이트 단어들의 출현 빈도를 사전형태로 리턴
    source = urllib2.urlopen(source).read() #source로 소스코드를 모두 문자열로 가져옴
    #영어 불용어 리스트
    stopword = [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']

    for i in range(0, source.count('<!--')): #주석태그부분 처리
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    for i in range(0,source.count("<")): #태그부분 처리
        if source.find("<") != -1:
            source = source.replace(source[source.find("<"):source.find(">")+1:],"")

    words = source.split() #words 리스트에 단어추출

    pure_words = []

    for s in words: #words 원소의 문자열을 하나하나씩 꺼내어 구두문자 제거
        for x in s: #문자열의 문자 하나하나씩 꺼내서 구두문자인지 비교
            if(x in string.punctuation): #문자가 구두문자이면 그 문자 제거
                s = s[:s.index(x)]+s[s.index(x)+1:]
        
        #구두문자를 제거했는데 빈문자열이면 리스트에 추가하지 않는다
        #영어 불용어면 추가하지 않는다
        if(s != '' and s not in stopword): 
            pure_words.append(s)


    D = {} #출현빈도를 저장할 사전

    for x in pure_words: #구두문자를 제거한 단어 하나하나씩 꺼내여 사전에 저장
        if(x in D): #단어가 key값에 이미 존재하면 value값에 1을 더한다.
            D[x] += 1
        else: #단어가 key값에 존재하지 않으면 value값을 1로하여 key를 추가한다.
            D[x] = 1
    return D

#각 URL로 지정된 웹페이지의 HTML 소스를 파일로 저장하는 함수
def URL_html(source):
    html = urllib2.urlopen(source).read()#html로 소스코드를 모두 문자열로 가져옴
   
    #URL중 http://를 지우고 www.부터를 file_name에 저장한다 
    if(source[:7]=="http://"): #http://로 시작하면 http://를 지운다
        file_name = source[7:]+".html"
    elif(source[:8]=="https://"): #https://로 시작하면 https://를 지운다
        file_name = source[8:]+".html"
    else: #http://가 아니면 바로 저장
        file_name = source+".html"
    
    #html 파일을 생성
    f = open(file_name,'w')
    f.write(html)
    f.close()

#단어의 출현빈도가 담긴 사전 객체를 위 HTML 소스 파일과 동일한 폴더에 파일로 저장하는 함수
def dict_store(source):
    
    #URL중 http://를 지우고 www.부터를 file_name에 저장한다 
    if(source[:7]=="http://"): #http://로 시작하면 http://를 지운다
        file_name = source[7:]+".words.frequency.txt"
    elif(source[:8]=="https://"): #https://로 시작하면 https://를 지운다
        file_name = source[8:]+".words.frequency.txt"
    else: #http://가 아니면 바로 저장
        file_name = source+".words.frequency.txt"

    #파일을 생성
    f = open(file_name,'w')
    pickle.dump(extraction_dictionary(source),f) #pockle 모듈 사용
    f.close()

#words.frequency 파일을 객체로 다시 리턴하는 함수
def object_load(source):
    
    #URL중 http://를 지우고 www.부터를 file_name에 저장한다 
    if(source[:7]=="http://"): #http://로 시작하면 http://를 지운다
        file_name = source[7:]+".words.frequency.txt"
    elif(source[:8]=="https://"): #https://로 시작하면 https://를 지운다
        file_name = source[8:]+".words.frequency.txt"
    else: #http://가 아니면 바로 저장
        file_name = source+".words.frequency.txt"
    
    f = open(file_name)
    return pickle.load(f) #파일을 객채로 리턴   
    
#영어 웹사이트 5개 html파일과 words.frequency파일 생성
URL_html('http://thenextweb.com')
dict_store('http://thenextweb.com')

URL_html('https://github.com')
dict_store('https://github.com')

URL_html('http://mashable.com')
dict_store('http://mashable.com')

URL_html('http://techcrunch.com')
dict_store('http://techcrunch.com')

URL_html('http://recode.net')
dict_store('http://recode.net')

#pickle 모듈로 저장한 5개의 words.frequency를 다시 5개의 사전 객체로 로드 하는 코드
d1 = object_load('http://thenextweb.com')
d2 = object_load('https://github.com')
d3 = object_load('http://mashable.com')
d4 = object_load('http://techcrunch.com')
d5 = object_load('http://recode.net')

D = {} #5개의 사전에 있는 단어들을 하나로 통합할 사전
item_list = []

#하나로 통합하기 위해서 사전을 리스트로 반환
item_list.append(d1.items())
item_list.append(d2.items())
item_list.append(d3.items())
item_list.append(d4.items())
item_list.append(d5.items())

#리스트에서 (words,values)로 되어있는 튜플을 하나씩 확인하여 단어 빈도를 D 사전에 저장
for x in item_list: #리스트에서 한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트 하나씩 꺼냄 
    for words in x: #한 웹페이지의 단어빈도가 튜플형태로 저장되어 있는 리스트에서 하나의 튜플을 꺼냄
        if(words[0] in D): #(단어, 빈도)로 저장되어있는 튜플 words에서 단어가 D 사전에 이미 존재하면 빈도를 +1
            D[words[0]] += 1
        else: #단어가 존재하지 않으면 빈도를 1로하여 추가한다.
            D[words[0]] = 1

#sort하기 위하여 D 사전을 리스트로 변환
result_list = D.items()

#단어의 빈도(사전의 value)값으로 정렬
result_list.sort(key=lambda t:t[1], reverse=True)

#가장 많이 출현한 단어 3개 (단어,출현한 횟 수) 형태로 출력 
print result_list[0]
print result_list[1]
print result_list[2]
```

함수 개요
* extraction_dictionary(source)
    * 주소를 적으면 해당 사이트 단어들의 출현 빈도를 사전형태로 리턴하는 함수
* URL_html(source)
    * 각 URL로 지정된 웹페이지의 HTML 소스를 파일로 저장하는 함수
* dict_store(source)
    * 단어의 출현빈도가 담긴 사전 객체를 위 HTML 소스 파일과 동일한 폴더에 파일로 저장하는 함수
* object_load(source)
    * words.frequency 파일을 객체로 다시 리턴하는 함수
